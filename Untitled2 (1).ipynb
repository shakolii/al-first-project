```python
import numpy as np
from matplotlib import pyplot
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
x = np.random.rand(5,1) *10 -5
y = 2*x +5
y = y + np.random.normal(scale =1 , size = y.shape)
plt.plot(x,y, "r")
```

```python
poly_features = PolynomialFeatures(degree=1 )
x_poly = poly_features.fit_transform(x)
reg1_1= LinearRegression()
reg1_1.fit(x_poly , y)
```

```python
x_vals = x
x_vals_poly = poly_features.transform(x_vals)
y_vals = reg1_1.predict(x_vals_poly)
plt.suptitle('DATA 1 : Degree 1')
plt.ylabel('Output')
plt.xlabel('Input')
plt.ylim((-20,20))
plt.xlim((-5,5))
plt.scatter(x, y)
plt.plot(x_vals, y_vals, color="r")
plt.show()
```

```python
print(f" intercept : {reg1_1.intercept_} \n coefficient : {reg1_1.coef_} \n MSE : {metrics.mean_squared_error(y, y_vals)} \n R2 Score : {r2_score(y, y_vals)}")
```

```python
poly_features_4 = PolynomialFeatures(degree=4, include_bias=False)
x_poly_4 = poly_features_4.fit_transform(x)
reg1_1_4 = LinearRegression()
reg1_1_4.fit(x_poly_4, y)
```

```python
x_vals_4_show = np.linspace(-5,5,500).reshape(-1,1)
x_vals_poly_4_show = poly_features_4.transform(x_vals_4_show)
y_vals_4_show = reg1_1_4.predict(x_vals_poly_4_show)
x_vals_4 = x
x_vals_poly_4 = poly_features_4.transform(x_vals_4)
y_vals_4 = reg1_1_4.predict(x_vals_poly_4)
plt.suptitle('DATA 1 : Degree 4')
plt.ylabel('Output')
plt.xlabel('Input')
plt.ylim((-20,20))
plt.xlim((-5,5))
plt.scatter(x, y)
plt.plot(x_vals_4_show, y_vals_4_show, color="r")
plt.show()
```

```python
print(f" intercept : {reg1_1_4.intercept_} \n coefficient : {reg1_1_4.coef_} \n MSE : {metrics.mean_squared_error(y, y_vals_4)} \n R2 Score : {r2_score(y, y_vals_4)}")
```
